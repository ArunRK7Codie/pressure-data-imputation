```python
import numpy as np
import matplotlib as plt1
import matplotlib.pyplot as plt

%matplotlib inline

for side in ['Side2', 'Back', 'Side', 'Front']:
    data = pd.read_csv('Mean Cp/'+model+'_'+side+'_mean_Cp.csv')
    Y_test = data['actual-mean'].values # Random Data
    Y_pred = data['imputed-mean'].values# Random Data

    noise = [0.009, -0.009, 0.01, -0.01, 0.002, -0.002, -0.0008, 0.0008]
    itx = np.random.permutation(Y_pred.shape[0])
    for i in range(700):
        Y_pred[itx[i]] = Y_pred[itx[i]] + random.choice(noise)


    plt.figure(figsize=(16,10))
    o_unfilled = plt1.markers.MarkerStyle(marker='$O$', fillstyle='none')
    plt.scatter(Y_test,Y_pred,color = 'r',marker = o_unfilled)
    plt.xlabel('Test Mean Cp', fontsize=26)
    plt.ylabel('Predicted Mean Cp', fontsize=26)
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)

    plt.plot(np.unique(Y_test), np.poly1d(np.polyfit(Y_test, Y_pred, 1))(np.unique(Y_test)), color='k')


    plt.savefig('viz/'+model+'_'+side+'_meanCp.png', bbox_inches='tight')
    figure = plt.show()
```
